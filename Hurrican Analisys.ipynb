```python
def damages_float(damages):
    damages_updated =  []
    for damage in damages:
        if damage[-1] =="B":
            value = float(damage[:-1])
            damages_updated.append(value*1000000000)
        elif damage[-1] =="M":
            value = float(damage[:-1])
            damages_updated.append(value*1000000)
        else:
            damages_updated.append(damage)
    return damages_updated
```

```python
damages_updated = damages_float(damages)
damages_updated
```

```python
def areas_affected_count(areas_affected):
    areas_affected_dic = {}
    for areas in areas_affected:
        for area in areas:
            if area not in areas_affected_dic.keys():
                areas_affected_dic[area] = 1
            else:
                areas_affected_dic[area] += 1
    return areas_affected_dic
```

```python
areas_affected_by_count = areas_affected_count(areas_affected)
areas_affected_by_count
```

```python
def find_area_most_affected(areas_affected_by_count):
    most_affected_area = ['', 0]
    for area in areas_affected_by_count.keys():
        if areas_affected_by_count[area] > most_affected_area[1]:
            most_affected_area = [area, areas_affected_by_count[area]] 
    return most_affected_area
```

```python
area_most_affected_by_hurricans = find_area_most_affected(areas_affected_by_count)
area_most_affected_by_hurricans
```

```python
def find_greatest_number_of_deaths(names, deaths):
    greatest_number_of_deaths = ["", 0]
    for i in range(len(deaths)):
        if deaths[i] > greatest_number_of_deaths[1]:
            greatest_number_of_deaths = [names[i], deaths[i]]
    return greatest_number_of_deaths
```

```python
mortality_scale = {0: 0,
                   1: 100,
                   2: 500,
                   3: 1000,
                   4: 10000}
def assign_mortality_rate(names, deaths): 
    hurricans_by_mortality_rate = {0: [], 1: [], 2: [], 3: [], 4: [] }    
    for i in range(len(deaths)):
        if deaths[i] > mortality_scale[4]:
            hurricans_by_mortality_rate[4].append(names[i])
        elif deaths[i] > mortality_scale[3]:
            hurricans_by_mortality_rate[3].append(names[i])
        elif deaths[i] > mortality_scale[2]:
            hurricans_by_mortality_rate[2].append(names[i])
        elif deaths[i] > mortality_scale[1]:
            hurricans_by_mortality_rate[1].append(names[i])
        elif deaths[i] > mortality_scale[0]:
            hurricans_by_mortality_rate[0].append(names[i])
    return hurricans_by_mortality_rate
            
```

```python
hurricans_with_mortality_rate = assign_mortality_rate(names, deaths)
hurricans_with_mortality_rate
```
